# Image Threshold

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv.imread('../assets/rainbow.jpg')
plt.imshow(img)
```

## Importing as Grayscale

```python
img = cv.imread('../assets/rainbow.jpg', 0)
plt.imshow(img, cmap="gray")
```

## Threshold

```python
img.max()
```

```python
ret, thresh_img = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
plt.imshow(thresh_img, cmap="gray")
```

```python
img = cv.imread('../assets/crossword.jpg', 0)
plt.imshow(img, cmap="gray")
```

```python
def show_pic(img, cmap):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap=cmap)
```

```python
show_pic(img, "gray")
```

### Simple Threshold

```python
ret, simple_thresh_img = cv.threshold(img, 180, 255, cv.THRESH_BINARY)
show_pic(simple_thresh_img, "gray")
```

### Adaptive Threshold

```python
block_size = 11
c = 8

adaptive_thresh_img = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, block_size, c)
show_pic(adaptive_thresh_img, "gray")
```

### Blending Threshold

```python
blended_img = cv.addWeighted(src1=simple_thresh_img, alpha=0.8,src2=adaptive_thresh_img, beta=0.1, gamma=0)
show_pic(blended_img, "gray")
```
