# OpenCV Drawing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

## Rectangle

```python
cv2.rectangle(blank_img,pt1=(100,10), pt2=(500,150), color=(0,255,0), thickness=5)
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=5)
plt.imshow(blank_img)
```

## Circle

```python
cv2.circle(blank_img, center=(100,100), radius=50, color=(255,0,0), thickness=5)
plt.imshow(blank_img)
```

_thickness = -1_ means that the shape will be filled

```python
cv2.circle(blank_img, center=(400,400), radius=50, color=(255,0,0), thickness=-1)
plt.imshow(blank_img)
```

## Line

```python
cv2.line(blank_img, pt1=(0,0), pt2=(512,512), color=(255,255,0), thickness=1)
plt.imshow(blank_img)
```

## Text

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text="hello", org=(10,500), fontFace=font, fontScale=4, color=(255,255,255), thickness=3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

## Polygon

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int32)
```

```python
vertices = np.array([[100,100], [100,200], [200,200], [300,200]], np.int32)
vertices.shape
```

```python
vertices
```

Cv2 requires an specific array format with 3 dimentions, we created a 4x2, so we need to reshape using (-1,1,2) usually

```python
reshaped_vertices = vertices.reshape((-1,1,2))
reshaped_vertices.shape
```

```python
reshaped_vertices
```

```python
cv2.polylines(blank_img, [reshaped_vertices], isClosed=True, color=(255,0,0), thickness=5)
plt.imshow(blank_img)
```
