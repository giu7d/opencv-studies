# Basics of OpenCV
## Setting Up

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

## Importing Images

```python
img = cv2.imread("assets/DATA/00-puppy.jpg")
type(img)
```

```python
img.shape
```

### Images representation

OpenCV represent images in a diferent shape than matplotlib.

#### Representation:

__MatPlotLib:__ Red Green Blue (RGB)
__OpenCV:__ Blue Green Red (BGR)

```python
#Matplotlib
plt.imshow(img)
```

```python
# Converting CV2 to PLT
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

### Importing as grayscale

```python
img_gray = cv2.imread("assets/DATA/00-puppy.jpg", cv2.IMREAD_GRAYSCALE)
img_gray.shape
```

```python
plt.imshow(img_gray, cmap="gray")
```

## Modifing Images
### Resize

```python
small_img = cv2.resize(rgb_img,(1000, 500))
plt.imshow(small_img)
```

```python
w_ratio = 0.05
h_ratio = 0.05

resize_img = cv2.resize(rgb_img, (0,0), rgb_img, w_ratio, h_ratio)
plt.imshow(resize_img)
```

### Flip

```python
flip_img = cv2.flip(rgb_img, 0)
plt.imshow(flip_img)
```

```python
flip_img = cv2.flip(rgb_img, 1)
plt.imshow(flip_img)
```

```python
flip_img = cv2.flip(rgb_img, -1)
plt.imshow(flip_img)
```

## Saving Images

```python
cv2.imwrite('flipped_dog.jpg', flip_img)
```
