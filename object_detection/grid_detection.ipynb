# Grid Detection

Grid patterns are often used to calibrate cameras and object tracking.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Square Grid

```python
chess_img = cv.imread("../assets/flat_chessboard.png")
plt.imshow(chess_img)
```

```python
found, corners = cv.findChessboardCorners(chess_img, (7,7))
```

```python
chess_img_copy = chess_img.copy()
cv.drawChessboardCorners(chess_img_copy, (7,7), corners, found)
plt.imshow(chess_img_copy)
```

## Circle Grid

```python
circles_img = cv.imread("../assets/dot_grid.png")
plt.imshow(circles_img)
```

```python
found, corners = cv.findCirclesGrid(circles_img, (10,10), cv.CALIB_CB_SYMMETRIC_GRID)
```

```python
circles_img_copy = circles_img.copy()
cv.drawChessboardCorners(circles_img, (10,10), corners, found)
plt.imshow(circles_img)
```
